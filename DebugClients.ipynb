```python
import macrodata as md
import pandas as pd
import requests
import json

from datetime import date
from datetime import datetime as dt

from macrodata.clients.inegi.utils import ticker_map, EoXMonth
from macrodata.clients.inegi.utils_private import API_KEYS
```

```python
tickers = md.Explorer.countryList(['Mexico'])['Mexico']
```

```python
tickers
```

```python
start_dt = date(1970, 1, 1)
end_dt = date(2022, 11, 1)
```

```python
def _get_inegi(tickers:list[str], start_dt, end_dt) -> pd.DataFrame:
        _data = []
        p = {'type': 'json'}
        
        for tic in tickers:
            url = ('http://en.www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/'
            + ticker_map[tic]['id'] 
            + '/en/0700/false/' 
            + ticker_map[tic]['fuente'] 
            + '/2.0/' 
            + API_KEYS['INEGI_API_KEY'])
            s = requests.get(url, p).text
            _d = pd.DataFrame(json.loads(s)['Series'][0]['OBSERVATIONS']).set_index('TIME_PERIOD')
            _d.index = _d.index.map(lambda x: EoXMonth(dt.strptime(x, '%Y/%m').date(), 0, False))
            _d = _d.loc[_d.index >= start_dt].sort_index()['OBS_VALUE']
            _d.name = tic
            _data.append(pd.DataFrame(_d))

        _data = _data[0].join(_data[1:], how='outer')
        return _data
```
