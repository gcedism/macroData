```python
import sqlalchemy as sa
import sqlalchemy.orm as orm
from sqlalchemy.engine.url import URL
```

```python
DATABASE = {
    'drivername': 'sqlite',
    'database': 'macrodata/_data/macroDatabase.db'
}
engine = sa.create_engine(URL.create(**DATABASE))
Base = orm.declarative_base()
    
class History(Base) :
    __table__ = sa.Table('monthly_history', Base.metadata, autoload_with=engine)
    
    def __repr__(self) :
        return f'History({self.index})'

class Data(Base) :
    __table__ = sa.Table('data_specs', Base.metadata, autoload_with=engine)
    
    def __repr__(self) :
        return f'Data({self.id} from {self.source})'
    
session = orm.Session(engine)
```

```python
items = session.scalars(sa.select(History)).all()
```

```python
items[0].eu_ind_conf
```

```python
from datetime import timedelta as td
isRealEndOfMonth = lambda x : (x+td(days=1)).month != x.month
```

```python
toDelete = [item for item in items if isRealEndOfMonth(item.index) == False]
```

```python
{toDelet.index :[(col.name, toDelet.__dict__[col.name])
 for col in toDelet.__table__.columns if toDelet.__dict__[col.name] != None]
 for toDelet in toDelete}
```

```python
[x.index for x in toDelete]
```

```python
cols = [History.index] + [col for col in History.__table__.c if col.name in items]
stmt = (sa.select(*cols)
    .filter(History.__table__.c[items[0]] != None)
    .order_by(History.index.desc())
    .limit(1))
results = engine.execute(stmt)
lastHistorySource = results.first().index
```
